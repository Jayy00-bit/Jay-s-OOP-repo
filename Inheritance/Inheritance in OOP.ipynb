```python
#Example 1 : Simple Inheritance
class Person:
    def greet(self):
        print("Good day Everyone!")

class Student(Person):
    pass

s = Student()
s.greet()
```

```python
#Example 2 : Inheriting Attributes

class Person:
    def __init__(self,name):
        self.name = name

class Student(Person):
    pass

s = Student("Matthew")
print(s.name)
```

```python
#Example 3 : Extending a Parent Class

class Person:
    def greet(self):
        print("Hello")

class Student(Person):
    def study(self):
        print("I'm Enjoying the Python OOP Course")

Daniella = Student()

Daniella.greet()
Daniella.study()
```

```python
#Example 4a :Constructor Inheritance
class Person:
    def __init__(self,name):
        self.name = name
class Student(Person):
    def __init__(self,name,student_id):
        Person.__init__(self,name)
        self.student_id = student_id

student = Student("Matthew","S12345")
print(student.name)
print(student.student_id)


#Example 4b
class Person:
    def __init__(self,name):
        self.name = name

class Student(Person):
    def __init__(self,name,student_id):
        super().__init__(name)
        self.student_id = student_id

student = Student("Matthew","S12345")
print(student.name)
print(student.student_id)
```

```python
#Example 5

class Person:
    def __init__(self,name):
        self.name = name

    def display(self):
        print("Name:", self.name)

class Student(Person):
    def __init__(self,name,student_id):
        super().__init__(name)
        self.student_id = student_id

    def info(self):
        print("ID:", self.student_id)

student = Student("John","1001")

student.display()
student.info()
```

```python
#Method Overriding
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

dog = Dog()
dog.speak()
```

```python
#Calling the Parent method

class Animal():
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")

dog = Dog()
dog.speak()
```

```python
#Method Resolution Order (MRO)

class A:
    def show(self):
        print("A")

class B(A):
    pass

class C(A):
    pass

class D(B,C):
    pass

help(D)
```

```python
#Lab Session
#Case Study Activity: UMaT Management System
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name:{self.name}, Age:{self.age}")

class Student(Person):
    def __init__(self,name,age,student_id,enroll_course):
        super().__init__(name,age)
        self.student_id = student_id
        self.enroll_course = enroll_course

    def display_info(self):
        print(f"This is {self.name} \nHe is {self.age} \nHe has Student ID {self.student_id} \nCourse Enrollment Status = {self.enroll_course}")

class Lecturer(Person):
    def __init__(self,name,age,employee_id,teach_course):
        super().__init__(name,age)
        self.employee_id = employee_id
        self.teach_course = teach_course

    def show_more_info(self):
        print(f"Here stands a Honorable Lecturer with Employee_id: {self.employee_id} and Course Teaching Status: {self.teach_course} ")


student = Student("Jayy",18,9016915325,True)
student.display_info()

lecturer = Lecturer("Dr.Matthew Cobbinah",40,"12/06/89",True)
lecturer.display_info()
lecturer.show_more_info()
```
